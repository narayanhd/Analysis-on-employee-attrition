    What are the factors that contribute more to employee attrition?
    What measures should company take in order to retain their employees?
 Who is leaving? When they are leaving? Why they are leaving?
 
 Objective: To strike the right balance of holding on to top talent while accepting that some level of attrition is healthy
 
 1st question: What is Employee attrition? 
 Answer: It is a naturally occurring, voluntary departure of employees from a company.
 Employee attrition means leaving a job for different reasons: personal, professional motivation, job mismatch....

#### Age Factor

```python
# show the number of employees that left and stayed by age
plt.subplots(figsize=(12,4))
sns.countplot(x='Age',hue='Attrition',data=df,palette='pink')

# yes = left company
# no = stayed
```

```python
def age_group(Age):
    if Age <=25:
        return '18-25'
    if Age <=35:
        return '26-35'
    if Age <=45:
        return '36-45'
    if Age <=55:
        return '46-55'
    else:
        return 'over 55'
    
df['AgeGroup']=df.Age.apply(age_group)
df.head()
```

```python
# Attrition_by_age_group ( Distribution of employees that left by Age Group )
attrition_counts=df[df['Attrition']=='Yes'].groupby('AgeGroup').size()
total_counts=df.groupby('AgeGroup').size()
attrition_counts
```

```python
df['AgeGroup'].value_counts()
```

```python
attrition_rates=(attrition_counts/total_counts)*100
attrition_rates
```
